```python
import numpy as np
import pandas as pd
import pickle

# Load the data from saved pickle files
popular_df = pickle.load(open('popular.pkl', 'rb'))
pt = pickle.load(open('pt.pkl', 'rb'))
books = pickle.load(open('books.pkl', 'rb'))
similarity_scores = pickle.load(open('similarity_scores.pkl', 'rb'))

def recommend(book_name=None, genre=None):
    if book_name:
        # Book-based recommendation
        index = np.where(pt.index == book_name)[0][0]
        similar_items = sorted(list(enumerate(similarity_scores[index]), key=lambda x: x[1], reverse=True)[1:5])

        data = []
        for i in similar_items:
            item = []
            temp_df = books[books['Book-Title'] == pt.index[i[0]]]
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
            item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
            data.append(item)

        return data
    elif genre:
        # Genre-based recommendation
        genre_books = popular_df[popular_df['Genre'].str.lower() == genre.lower()]
        if genre_books.empty:
            return "No popular books found in the specified genre."
        return genre_books[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_rating']].values.tolist()
    else:
        return "Please provide a book name or genre for recommendations."

def main():
    print("Choose an option:")
    print("1. Recommend by Book Name")
    print("2. Recommend by Genre")
    
    user_choice = input("Enter your choice (1/2): ")
    
    if user_choice == '1':
        book_name = input("Enter the book name: ")
        book_recommendations = recommend(book_name=book_name)
        if book_recommendations:
            print("Book Recommendations:")
            for book in book_recommendations:
                print(book)
        else:
            print("No recommendations found for the specified book.")
    elif user_choice == '2':
        user_genre = input("Enter a genre for popular book recommendations: ")
        genre_recommendations = recommend(genre=user_genre)
        if isinstance(genre_recommendations, str):
            print(genre_recommendations)
        else:
            print("Genre Recommendations:")
            for book in genre_recommendations:
                print(book)
    else:
        print("Invalid choice. Please enter 1 or 2.")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python
# coding: utf-8

import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def main():
    books = pd.read_csv('books.csv')
    users = pd.read_csv('users.csv')
    ratings = pd.read_csv('ratings.csv')

    ratings_with_name = ratings.merge(books, on='ISBN')
    
    num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
    num_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)

    # Convert the 'Book-Rating' column to numeric, coercing non-numeric values to NaN
    ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')
    
    avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
    avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
    
    popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
    
    books_with_genre = pd.read_csv('Books.csv')
    popular_df = popular_df.merge(books_with_genre, on='Book-Title').drop_duplicates('Book-Title')
    
    popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_rating', ascending=False).head(50)

    x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
    well_ware_users = x[x].index

    filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(well_ware_users)]

    y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
    famous_books = y[y].index

    final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]

    pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
    pt.fillna(0, inplace=True)

    similarity_scores = cosine_similarity(pt)

    while True:
        print("Choose an option:")
        print("1. Recommend by Book Name")
        print("2. Recommend by Genre")
        print("3. Quit")
        
        user_choice = input("Enter your choice (1/2/3): ")
        
        if user_choice == '1':
            user_book = input("Enter a book name for recommendations: ")
            book_recommendations = recommend(book_name=user_book)
            if book_recommendations:
                print("Recommended Books:")
                for rec in book_recommendations:
                    print(f"Title: {rec[0]}")
                    print(f"Author: {rec[1]}")
                    print(f"Image URL: {rec[2]}")
                    print()
            else:
                print("Book not found or insufficient data for recommendations.")
        
        elif user_choice == '2':
            user_genre = input("Enter a genre for popular book recommendations: ")
            genre_recommendations = recommend(genre=user_genre)
            if isinstance(genre_recommendations, str):
                print(genre_recommendations)
            else:
                print("Recommended Books:")
                for rec in genre_recommendations.iterrows():
                    print(f"Title: {rec[1]['Book-Title']}")
                    print(f"Author: {rec[1]['Book-Author']}")
                    print(f"Image URL: {rec[1]['Image-URL-M']}")
                    print()

        elif user_choice == '3':
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

def recommend(book_name=None, genre=None):
    if book_name:
        # Recommendation based on book name
        if book_name not in pt.index:
            return "Book not found in the dataset."

        index = np.where(pt.index == book_name)[0][0]
        similar_items = sorted(list(enumerate(similarity_scores[index]), key=lambda x: x[1], reverse=True)[1:5])

        data = []
        for i in similar_items:
            item = []
            temp_df = books[books['Book-Title'] == pt.index[i[0]]]
            item.extend(list(temp_df[temp_df['Book-Title'].duplicated(keep='first')]['Book-Title'].values))
            item.extend(list(temp_df[temp_df['Book-Title'].duplicated(keep='first')]['Book-Author'].values))
            item.extend(list(temp_df[temp_df['Book-Title'].duplicated(keep='first')]['Image-URL-M'].values))

            data.append(item)

        return data
    elif genre:
        # Genre-based recommendation
        genre_books = popular_df[popular_df['Genre'].str.lower() == genre.lower()]
        if genre_books.empty:
            return "No popular books found in the specified genre."
        genre_books = genre_books.sort_values('avg_rating', ascending=False).head(5)
        return genre_books

if pt.shape[0] == 0:
    print("The pivot table (pt) is empty. Please ensure you have loaded data correctly.")
else:
    similarity_scores = cosine_similarity(pt)




if __name__ == "__main__":
    main()
```
